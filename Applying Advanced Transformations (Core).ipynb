## Hero Info

```python
## Load in csv.gz
info = pd.read_csv('Data/superhero_info.csv')
info.head()
```

```python
info.duplicated().sum()
```

```python
info.isna().sum()
```

### Split Hero and Publisher to separate columns

```python
## adding expand=True
info['Hero|Publisher'].str.split('|',expand=True)
```

```python
## save the 2 new columns into the dataframe
info[['Hero_names','Publisher']] = info['Hero|Publisher'].str.split('|',expand=True)
info.head(2)
```

```python
## drop the original column 
info = info.drop(columns=['Hero|Publisher'])
info.head(2)
```

### Split Height and Weight into separate columns

```python
## use .apply pd.Series to convert a dict to columns
info['Measurements'].apply(pd.Series)
```

```python
## replace ' with " (entire column)
info['Measurements'] = info['Measurements'].str.replace("'", '"')
## apply json.loads
info['Measurements'] = info['Measurements'].apply(json.loads)
```

```python
## use .apply pd.Series to convert a dict to columns
info['Measurements'].apply(pd.Series)
```

```python
## Concatenate the 2 new columns and drop the original.
info = pd.concat([info, info['Measurements'].apply(pd.Series)], axis = 1)
#df = df.drop(columns = 'coordinates')
info.head(2)
```

```python
## drop the original column 
info = info.drop(columns=['Measurements'])
info.head(2)
```

```python
## adding expand=True
info['Height'].str.split(' ',expand=True)
```

```python
## save the 2 new columns into the dataframe
info[['Height_cm','cm']] = info['Height'].str.split(' ',expand=True)
info.head(2)
```

```python
## drop the original column 
info = info.drop(columns=['Height', 'cm'])
info.head(2)
```

```python
## adding expand=True
info['Weight'].str.split(' ',expand=True)
```

```python
## save the 2 new columns into the dataframe
info[['Weight_kg','kg']] = info['Weight'].str.split(' ',expand=True)
info.head(2)
```

```python
info = info.drop(columns = ['Weight', 'kg'])
info.head(2)
```

```python
datatypes = info.dtypes['Weight_kg']
datatypes
```

```python
# convert just columns Weight_kg and Height_cm
info[["Weight_kg","Height_cm"]] = info[["Weight_kg","Height_cm"]].apply(pd.to_numeric)
info
```

```python
datatypes = info.dtypes['Weight_kg']
datatypes
```

```python
info.info()
```

```python
## save data for next lesson
info.to_csv('advanced_hero_info.csv', index=False)
```

### Powers 

```python
power = pd.read_csv('Data/superhero_powers.csv')
power.head()
```

```python
power = power.rename(columns ={'hero_names':'Hero_names'})
power.head()
```

```python
## showing the lists are really strings
power.loc[2,'Powers']
```

```python
## adding expand=True
power['power_split'] = power['Powers'].str.split(',')
```

```python
## exploding the column of lists
exploded = power.explode('power_split')
exploded.head()
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['power_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    power[col] = power['Powers'].str.contains(col)
power.head()
```

```python
# drop transactions clumns
power_f = power.drop(columns=['Powers','power_split'])
power_f.head()
```

```python
## save data for next lesson
power_f.to_csv('advanced_power.csv', index=False)
```

```python
new_df = pd.concat([info, power_f], axis=1)
new_df.head()
```

```python
## save data for next lesson
new_df.to_csv('advanced_hero_power.to_csv', index=False)
```
